### decompose the flag into new binary df

```python
flag_arr = df.qualityFlag
bin_values = np.flip(2 ** np.arange(11),0)
flag_bin_arr = np.zeros([len(flag_arr),len(bin_values)])

for i, flag in enumerate(flag_arr):
    bin_str = "{:011b}".format(flag)
    bin_int = list(map(int,list(bin_str)))
    flag_bin_arr[i,:] = bin_int
```

```python
flagdf = df[['qualityFlag']].copy()
for i,bin_value in enumerate(bin_values):
    flagdf['{}'.format(bin_value)] = flag_bin_arr[:,i].astype(int)
```

```python
fig, axes = plt.subplots(2,4,figsize=(32,12))
axes = axes.flatten()

bin_values = (2 ** np.arange(8)).astype(str)

for i,col_name in enumerate(bin_values):
    idx0 = np.arange(len(df))[((df['mergedClass'] == 1) | (df['mergedClass'] == -1)) & (flagdf[col_name] == 0)]
    idx1 = np.arange(len(df))[((df['mergedClass'] == 1) | (df['mergedClass'] == -1)) & (flagdf[col_name] == 1)]
    
    axes[i].scatter(df.uk_mAB_I[idx0],df.yMeanPSFMag[idx0],c='b',s=10,alpha=0.3,label='flag {} = 0'.format(col_name))
    axes[i].scatter(df.uk_mAB_I[idx1],df.yMeanPSFMag[idx1],c='r',s=10,alpha=0.3,label='flag {} = 1'.format(col_name))
    axes[i].plot([min(df.yMeanPSFMag),max(df.yMeanPSFMag)],[min(df.yMeanPSFMag),max(df.yMeanPSFMag)],'k')
    axes[i].plot([min(df.yMeanPSFMag),max(df.yMeanPSFMag)],[min(df.yMeanPSFMag),max(df.yMeanPSFMag)],'k')
    axes[i].set_xlabel(r'UKIDSS $m_{AB}^I$ (1.05 um)',fontsize='xx-large')
    axes[i].set_ylabel(r'PanSTARRS  $m_{AB}^y$ (0.9633 um)',fontsize='xx-large')
    axes[i].legend(loc=0,frameon=False,fontsize='xx-large')

plt.savefig('mag_corr_flag_comp.png',dpi=256,bbox_inches='tight')
```
